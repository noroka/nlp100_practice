# 第2章: UNIXコマンド

popular-names.txtは，アメリカで生まれた赤ちゃんの「名前」「性別」「人数」「年」をタブ区切り形式で格納したファイルである．以下の処理を行うプログラムを作成し，popular-names.txtを入力ファイルとして実行せよ．さらに，同様の処理をUNIXコマンドでも実行し，プログラムの実行結果を確認せよ．

https://nlp100.github.io/ja/ch02.html

## 10. ファイルの行数カウント
行数をカウントせよ．確認にはwcコマンドを用いよ．

wc -l ファイル名
```
root@4542cb37189e:/home/tmp# wc -l popular-names.txt 
2780 popular-names.txt
```

## 11. 文字列置換　
タブ1文字につきスペース1文字に置換せよ．確認にはsedコマンド，trコマンド，もしくはexpandコマンドを用いよ．

sed -e 's/置換前/置換後/g'
```
root@4542cb37189e:/home/tmp# cat popular-names.txt | sed -e 's/\t/ /g'
Mary F 7065 1880
Anna F 2604 1880
Emma F 2003 1880
Elizabeth F 1939 1880
:
```

## 12.タブ区切りファイルから指定列抽出　
1列目をcol1.txtに，2列目をcol2.txtに保存

cut -f 列番号 < 入力ファイル >
```
root@4542cb37189e:/home/tmp# cut -f 1 < popular-names.txt > cut1.txt | cut -f 2 < popular-names.txt > cut2.txt
root@4542cb37189e:/home/tmp# ls
cut1.txt  cut2.txt  popular-names.txt
```

## 13.ファイル列方向結合
12で作ったcol1.txtとcol2.txtを結合し，元のファイルの1列目と2列目をタブ区切りで並べたテキストファイルを作成せよ．確認にはpasteコマンドを用いよ．

```
root@4542cb37189e:/home/tmp# paste cut1.txt cut2.txt 
Mary	F
Anna	F
Emma	F
Elizabeth	F
```

## 17.指定列からユニークな要素を抽出
1列目の文字列の種類（異なる文字列の集合）を求めよ．確認にはcut, sort, uniqコマンドを用いよ．

```
root@4542cb37189e:/home/tmp# cut -f 1 < popular-names.txt | sort | uniq
Abigail
Aiden
Alexander
Alexis
```

## 18.各行を3コラム目の数値の降順にソート
各行を3コラム目の数値の逆順で整列せよ（注意: 各行の内容は変更せずに並び替えよ）．確認にはsortコマンドを用いよ（この問題はコマンドで実行した時の結果と合わなくてもよい）．

```
root@4542cb37189e:/home/tmp# sort -n -r -k 3,3 -t $'\t' popular-names.txt 
Linda	F	99689	1947
Linda	F	96211	1948
James	M	94757	1947
Michael	M	92704	1957
```

## 19.各行の1コラム目の文字列の出現頻度を求め，出現頻度の高い順に並べる
各行の1列目の文字列の出現頻度を求め，その高い順に並べて表示せよ．確認にはcut, uniq, sortコマンドを用いよ．

uniq -c で重複要素のカウントできる
```
root@4542cb37189e:/home/tmp# cut -f 1 < popular-names.txt | sort | uniq -c | sort -r -k 1,1 -t ' '
    118 James
    111 William
    108 Robert
    108 John
```
