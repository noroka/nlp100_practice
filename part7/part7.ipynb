## 62. 類似度の高い単語10件

“United States”とコサイン類似度が高い10語と，その類似度を出力せよ．

```python
# w2vのサイズが大きすぎて実行できないので、軽量化する。
# https://kamo.hatenablog.jp/entry/2020/04/05/173810
fname = dst_dir + "googlenews_w2v.kv"
googlenews_w2v.wv.save(fname)
```

```python
# kvファイルの読み込み
from gensim.models import KeyedVectors

dst_dir = 'org/'
fname = dst_dir + "googlenews_w2v.kv"
kvs = KeyedVectors.load(fname,mmap='r')

# most_similarは使えなかった（クラッシュした）
# print(kvs.most_similar('United_States'))
# googlenews_w2v.most_similar('United_states')
```

```python
import numpy as np
from tqdm.notebook import tqdm as tqdm

def cos_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

v1 = kvs['United_States']
n_sims = 10
sim_vals = np.zeros(n_sims)
sim_words = np.zeros(n_sims)

for i in tqdm(range(0,len(kvs.index2word))):
    w = kvs.index2word[i]
    v2 = kvs[w]
    sim = cos_sim(v1, v2)
    sim_vals = np.append(sim_vals, sim)
    sim_words = np.append(sim_words, w)
    
    idx = np.argsort(sim_vals)[::-1]
    sim_vals = sim_vals[idx][0:9]
    sim_words = sim_words[idx][0:9]

print(sim_words)
print(sim_vals)
```

## 63. 加法構成性によるアナロジー

“Spain”の単語ベクトルから”Madrid”のベクトルを引き，”Athens”のベクトルを足したベクトルを計算し，そのベクトルと類似度の高い10語とその類似度を出力せよ．

```python
v1 = kvs['Spain'] - kvs['Madrid']

n_sims = 10
sim_vals = np.zeros(n_sims)
sim_words = np.zeros(n_sims)

for i in tqdm(range(0,len(kvs.index2word))):
    w = kvs.index2word[i]
    v2 = kvs[w]
    sim = cos_sim(v1, v2)
    sim_vals = np.append(sim_vals, sim)
    sim_words = np.append(sim_words, w)
    
    idx = np.argsort(sim_vals)[::-1]
    sim_vals = sim_vals[idx][0:9]
    sim_words = sim_words[idx][0:9]

print(sim_words)
print(sim_vals)
```

## 64. アナロジーデータでの実験

単語アナロジーの評価データをダウンロードし，vec(2列目の単語) - vec(1列目の単語) + vec(3列目の単語)を計算し，そのベクトルと類似度が最も高い単語と，その類似度を求めよ．求めた単語と類似度は，各事例の末尾に追記せよ．

```python
# ファイルのダウンロード
import os
import pprint
import time
import urllib.error
import urllib.request

def download_file(url, dst_path):
    try:
        with urllib.request.urlopen(url) as web_file, open(dst_path, 'wb') as local_file:
            local_file.write(web_file.read())
    except urllib.error.URLError as e:
        print(e)

def download_file_to_dir(url, dst_dir):
    download_file(url, os.path.join(dst_dir, os.path.basename(url)))

dst_dir = 'org/'
url = "http://download.tensorflow.org/data/questions-words.txt"

if not os.path.exists(dst_dir):
    os.mkdir(dst_dir)

download_file_to_dir(url, dst_dir)
```

```python
import numpy as np
import re
from tqdm.notebook import tqdm as tqdm

def cos_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

def most_similar(v1, kvs, restrict_vocab, topn):
    sims = np.array([cos_sim(v1, kvs[kvs.index2word[i]]) for i in range(0,restrict_vocab)])
    idx = sims.argsort()[::-1]
    topn_sims = sims[idx][0:topn]
    topn_words = [kvs.index2word[i] for i in idx[0:topn]]    
    return {'topn_words':topn_words, 'topn_sims':topn_sims}

# ファイル読み込み
fname = dst_dir + 'questions-words.txt'
with open(fname) as f:
    word_analogies = f.readlines()

out = [word_analogies[0]]
restrict_vocab = 100
topn = 1

for l in tqdm(range(1,len(word_analogies))):
    words = word_analogies[l].split()
    try:
        v1 = kvs[words[1]]
        v2 = kvs[words[0]]
        v3 = kvs[words[2]]
    except:
        continue   
    v = v1 - v2 + v3
    res = most_similar(v, kvs, restrict_vocab, topn)
    most_simword = res['topn_words'][0]
    most_simval = res['topn_sims'][0]
    updated_line = re.sub("\n","",word_analogies[l]) + " " + str(most_simword) + " " + str(most_simval)
    out.append(updated_line)

print(out[1:5])
```

## 65.アナロジータスクでの正解率

64の実行結果を用い，意味的アナロジー（semantic analogy）と文法的アナロジー（syntactic analogy）の正解率を測定せよ．

```python
correct = 0
for i in range(1,len(out)):
    line = out[i]
    if line.split()[4] == line.split()[3]:
        correct += 1

Accuracy = correct/len(out)
print(Accuracy)
```

The WordSimilarity-353 Test Collectionの評価データをダウンロードし，単語ベクトルにより計算される類似度のランキングと，人間の類似度判定のランキングの間のスピアマン相関係数を計算せよ．

```python
# ファイルの解凍
import zipfile
import re

tar_file = dst_dir + 'wordsim353.zip'

def extractfile(tar_file, dst_path = None):
    with zipfile.ZipFile(tar_file) as existing_zip:
        existing_zip.extractall(dst_path)

extractfile(tar_file, dst_dir)
```

```python
# ファイル読み込み
import pandas as pd
dst_dir = 'org/'
fname1 = dst_dir + 'set1.csv'
fname2 = dst_dir + 'set2.csv'
set1 = pd.read_csv(fname1)
set2 = pd.read_csv(fname2)
```

```python
from scipy import stats
import numpy as np

set1_word1 = set1['Word 1'].values
set1_word2 = set1['Word 2'].values
set1_simmean = set1['Human (mean)'].values

set2_word1 = set2['Word 1'].values
set2_word2 = set2['Word 2'].values
set2_simmean = set2['Human (mean)'].values


# Set1
sims = np.zeros(len(set1_word1))

for i in range(0,len(set1_word1)):
    sim = kvs.similarity(set1_word1[i], set1_word1[i])
    # print(set1_words[i], set2_words[i], sim)
    sims[i] = sim
    
rank_w2v = sims.argsort()[::-1]+1
rank_human = set1_simmean.argsort()[::-1]+1

print("set1_spearman_corr:",stats.spearmanr(rank_w2v, rank_human))

# Set2
sims = np.zeros(len(set2_word1))

for i in range(0,len(set2_word1)):
    sim = kvs.similarity(set2_word1[i], set2_word2[i])
    # print(set1_words[i], set2_words[i], sim)
    sims[i] = sim
    
rank_w2v = sims.argsort()[::-1]+1
rank_human = set2_simmean.argsort()[::-1]+1

print("set2_spearman_corr:",stats.spearmanr(rank_w2v, rank_human))
```

## 67.k-meansクラスタリング

国名に関する単語ベクトルを抽出し，k-meansクラスタリングをクラスタ数k=5として実行せよ．

```python
countries = []
# 国名の取得
import requests
from bs4 import BeautifulSoup

# Webページを取得して解析する
load_url = 'http://www13.plala.or.jp/bigdata/country.html'
html = requests.get(load_url)
soup = BeautifulSoup(html.content, "html.parser")
```

```python
# 国名一覧表を取得
import re

table_countries = soup.select("table[summary='世界の国一覧表（＋国名コード一覧）（アルファベット順）'] tr")

countries = []
for i in range(1,len(table_countries)-1):
    country = re.split('<tr><td>|</td><td>',str(table_countries[i]))[1]
    country_norm = re.sub(",","",country)
    country_norm = re.sub(" ","_",country_norm)
    countries.append(country_norm)
print(countries)
```

```python
# k-means用データ作成
dat = np.zeros([len(countries),300])
i = 0
countries_name = []
for c in countries:
    try:
        v = kvs[c]
        dat[i,:] = v
        countries_name.append(c)
        i += 1
    except:
        continue

feat = dat[0:i,:] # (countries, embedding_dim)
```

```python
# k-meansクラスタリング
# 参考：https://qiita.com/maskot1977/items/34158d044711231c4292
from sklearn.cluster import KMeans 
kmeans_model = KMeans(n_clusters=5, random_state=10).fit(feat)
labels = kmeans_model.labels_
print(labels)
```

```python
# 可視化
from sklearn.decomposition import PCA #主成分分析器
import matplotlib.pyplot as plt
%matplotlib inline

#主成分分析の実行
pca = PCA()
pca.fit(feat)

# データを主成分空間に写像 = 次元圧縮
feat_pca = pca.transform(feat)

# 第一主成分と第二主成分でプロットする
code_color = plt.rcParams['axes.prop_cycle'].by_key()['color']
colors = [code_color[l] for l in labels]

plt.figure(figsize=(20, 20))

for x, y, name in zip(feat_pca[:, 0], feat_pca[:, 1], countries_name):
    plt.text(x, y, name, alpha=0.8, size=12)
plt.scatter(feat_pca[:, 0], feat_pca[:, 1], color=colors, alpha=0.8)
plt.title("Principal Component Analysis")
plt.xlabel("The first principal component score")
plt.ylabel("The second principal component score")
plt.show()
```

## 68. Ward法によるクラスタリング

国名に関する単語ベクトルに対し，Ward法による階層型クラスタリングを実行せよ．さらに，クラスタリング結果をデンドログラムとして可視化せよ．

```python
# 参考：https://gomiba.co.in/blog/archives/705
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

# 階層型クラスタリングの実施
# ウォード法 x ユークリッド距離
linkage_result = linkage(feat, method='ward', metric='euclidean')

# クラスタ分けするしきい値を決める
threshold = 0.7 * np.max(linkage_result[:, 2])

# 階層型クラスタリングの可視化
plt.figure(num=None, figsize=(20, 9), dpi=200, facecolor='w', edgecolor='k')
dendrogram(linkage_result, labels=countries_name, color_threshold=threshold)
plt.show()

# クラスタリング結果の値を取得
clustered = fcluster(linkage_result, threshold, criterion='distance')
```
